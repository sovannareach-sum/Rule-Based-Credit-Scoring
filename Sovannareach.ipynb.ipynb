```python
import pandas as pd

# Load dataset
dataset_path = r'C:\Users\Admin\Documents\school stuff\Semester 8(Senior)\CS 342 Artificial Intelligent\Credit Score Classification Dataset.csv'
df = pd.read_csv(dataset_path)

# Check the columns and first few rows
print(df.columns)
print(df.head())
```

```python
# Rule-Based Credit Scoring Function
def rule_based_credit_score(row):
    if row['Income'] >= 100000 or row['Education'] in ["Doctorate", "Master's Degree"]:
        return "High"
    elif 50000 <= row['Income'] < 100000 and row['Education'] in ["Bachelor's Degree", "Master's Degree"]:
        return "Medium"
    elif row['Income'] < 50000 or row['Education'] == "High School Diploma":
        return "Low"
    else:
        return "Unknown"

# Apply Rules
df['Predicted_Credit_Score'] = df.apply(rule_based_credit_score, axis=1)

# Show Results
print(df[['Age', 'Income', 'Education', 'Predicted_Credit_Score']].head(10))

# Save the Output
output_path = r'C:\Users\Admin\Documents\school stuff\Semester 8(Senior)\CS 342 Artificial Intelligent\Rule_Based_Credit_Score_Output.csv'
df.to_csv(output_path, index=False)

print("\n✅ Rule-Based Credit Scoring Completed and Saved to CSV.")
```

1.Advantages of Rule-Based Machine Learning in Credit Assessment:

Simplicity and Interpretability
Rule-based models are easy to understand and interpret. Each decision is made through clear "if-then" rules, making it transparent for both developers and non-technical stakeholders, such as financial managers.

Quick Implementation
It does not require complex algorithms or heavy computational resources. Rule-based models can be implemented quickly, especially for small to medium-sized datasets.

Human Expertise Integration
These models can directly incorporate human knowledge and domain expertise into the decision-making process, which is particularly useful in credit assessment where experts often have specific guidelines.

2.Limitations of Rule-Based Machine Learning in Credit Assessment:

Lack of Flexibility
Rule-based models are static and cannot automatically learn patterns from new data. Every update or improvement requires manual modification of rules.

Scalability Issues
As the complexity of the problem increases, the number of rules can become very large, making the system hard to maintain and prone to errors.

Poor Generalization
Rule-based models may not perform well on unseen or unusual data, as they rely strictly on predefined conditions and cannot adapt to complex relationships hidden in the data.

Subjectivity and Bias
Since rules are manually designed, they may unintentionally reflect biases from the rule designer, which could lead to unfair credit assessments.
